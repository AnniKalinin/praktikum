### Сколько книг вышло после 1 января 2000 года?

```python
query_books = '''
    SELECT 
    COUNT(book_id) as books_count 
    FROM books
    WHERE publication_date :: timestamp > '2000-01-01' :: timestamp'''
```

```python
books = pd.io.sql.read_sql(query_books, con = engine)
print(books)
```

### Количество обзоров и средняя оценка для каждой книги

```python
query_authors = '''
        SELECT
            books.book_id AS book_id,
            books.title AS book_title,
            SQ2.mean_rating AS mean_rating,
            SQ2.num_rating AS num_rating,
            SQ3.num_review AS num_rewiev
        FROM books
        LEFT JOIN 
        (SELECT 
            ratings.book_id as book_id,
            AVG(ratings.rating) as mean_rating,
            COUNT(ratings.rating) as num_rating
        FROM ratings
        GROUP BY 
            book_id) 
            AS SQ2 ON books.book_id = SQ2.book_id
        LEFT JOIN 
        (SELECT
            reviews.book_id AS book_id,
            COUNT(reviews.review_id) AS num_review
        FROM reviews
        GROUP BY 
            book_id) 
            AS SQ3 ON books.book_id = SQ3.book_id
            '''
```

```python
books_with_rating = pd.io.sql.read_sql(query_authors, con = engine)
books_with_rating.head()
```

books_with_rating - таблица, где для каждой книги посчитано количество ревью, количество рейтинговых оценок и средний рейтинг

### Какое издательство выпустило наибольшее число книг толще 50 страниц?

```python
query_publishers = '''
    SELECT
        SQ1.publisher AS publisher,
        COUNT(SQ1.book_id) AS books_count
    FROM
        (SELECT 
            books.publisher_id AS publisher_id,
            books.book_id AS book_id,
            books.num_pages AS num_pages,
            publishers.publisher_id,
            publishers.publisher AS publisher
        FROM books
        LEFT JOIN publishers ON books.publisher_id = publishers.publisher_id
        WHERE 
            num_pages > 50) AS SQ1
    GROUP BY
    publisher
    ORDER BY 
    books_count DESC
    LIMIT
    10
    '''
publishers = pd.io.sql.read_sql(query_publishers, con = engine)
publishers.head()
```

Издательство, которое выпустило наибольшее число книг толще 50 страниц  - Penguin Books

### Автор с самой высокой средней оценкой книг (учитывая только книги с 50 и более оценками)

```python
query_ratings = '''
     SELECT 
     SUB5.author AS author,
     AVG(SUB5.mean_rating) AS mean_rating
     FROM 
     (SELECT
     books.book_id AS book_id,
     authors.author,
     SUB4.mean_rating 
     FROM books
     LEFT JOIN 
            (SELECT 
            ratings.book_id as book_id,
            AVG(ratings.rating) as mean_rating,
            COUNT(ratings.rating) as num_rating
            FROM ratings
            GROUP BY book_id
            HAVING 
            COUNT(ratings.rating) > 50
            ORDER BY 
            mean_rating DESC 
            ) 
            AS SUB4 ON SUB4.book_id = books.book_id
     LEFT JOIN authors ON books.author_id = authors.author_id) AS SUB5
     WHERE 
     mean_rating IS NOT NULL
     GROUP BY 
     author
     ORDER BY
     mean_rating DESC
     LIMIT 
     10
     
            
    '''
ratings = pd.io.sql.read_sql(query_ratings, con = engine)
ratings.head()
```

Автор с самой высокой средней оценкой книг J.K. Rowling/Mary GrandPré

### Cреднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
query_reviews = '''
    SELECT
    AVG(count)
    FROM
    (SELECT
        reviews.username,
        COUNT(reviews.review_id)
    FROM reviews
    WHERE reviews.username IN  
        (SELECT 
            ratings.username AS username
        FROM ratings
        GROUP BY username
        HAVING COUNT(ratings.rating_id) > 50)
    GROUP BY username) AS SUB6
            '''
reviews = pd.io.sql.read_sql(query_reviews, con = engine)
reviews
```

Cреднее количество обзоров от пользователей, которые поставили больше 50 оценок = 24
